### Choose one variable and plot that variable four different ways

```python
sns.set(style='darkgrid')
```

```python
movie['genre'].value_counts().plot(kind='bar');
```

```python
sns.countplot('genre', data=movie);
plt.xticks(rotation=90)
```

```python
sns.stripplot('genre', data=movie);
plt.xticks(rotation=90)

```

```python
sns.swarmplot('genre',data=movie) ;
plt.xticks(rotation=90)
```

### Choose two continuous variables, and plot them  three different ways

```python
sns.scatterplot(x= 'gross', y='budget', data=movie )
```

```python
sns.lineplot(x='gross', y='budget', data=movie)
```

```python
sns.lineplot(x='budget', y='gross', data=movie)
plt.xticks(rotation=90)
```

### Choose one continuous variable and one categorical variable and plot them six times

```python
sns.barplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```

```python
sns.stripplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```

```python
sns.boxplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```

```python
sns.pointplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```

```python
sns.violinplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```

```python
sns.swarmplot(x='genre', y='gross', data=movie)
plt.xticks(rotation=90)
```
